###Comparison of model trained params (All)

```python
import matplotlib.pyplot as plt

total_params = [14815044, 20124740, 2508868, 23966212, 22007588, 21262892, 18698308]
trainable_params = [100356, 100356, 250884, 401412, 204804, 401412, 376324]

non_trainable_params = [total - trainable for total, trainable in zip(total_params, trainable_params)]

model_names = ['VGG16', 'VGG19', 'MobileNetV2', 'ResNet50V2', 'InceptionV3', 'Xception', 'DenseNet201']

bar_width = 0.25
x_pos = [i for i, _ in enumerate(model_names)]

plt.bar(x_pos, total_params, color='blue', width=bar_width, label='Total Params')
plt.bar([i + bar_width for i in x_pos], trainable_params, color='orange', width=bar_width, label='Trainable Params')
plt.bar([i + bar_width*2 for i in x_pos], non_trainable_params, color='red', width=bar_width, label='Non-Trainable Params')

plt.xlabel('Model Names')
plt.ylabel('Number of Parameters')
plt.title('Total Params vs Trainable Params vs Non-Trainable Params')
plt.xticks([i + bar_width for i in x_pos], model_names, rotation=45)
plt.legend()
plt.show()
```

### Training time (Kaggle)

```python
import matplotlib.pyplot as plt

training_time_sec = [2556.01, 2680.04, 2426.18, 2330.11, 2394.58, 2514.76, 2786.37] # in seconds
model_names = ['VGG16', 'VGG19', 'MobileNetV2', 'ResNet50V2', 'InceptionV3', 'Xception', 'DenseNet201']
colors = ['orange', 'darkmagenta', 'limegreen', 'crimson', 'gold', 'indigo', 'deepskyblue']

# Convert training time from seconds to minutes
training_time_min = [time/60 for time in training_time_sec]

bar_width = 0.75
x_pos = [i for i, _ in enumerate(model_names)]

# Create a loop to add text labels on top of each bar
for i in range(len(model_names)):
    plt.text(x_pos[i], training_time_min[i], str(round(training_time_min[i], 1)) + ' min', ha='center', va='bottom')

plt.bar(x_pos, training_time_min, color=colors, width=bar_width)

plt.xlabel('Model Names')
plt.ylabel('Training Time (in minutes)')
plt.title('Training Time of Seven Models (Kaggle)')
plt.xticks(x_pos, model_names, rotation=45)
plt.show()
```

### Plotting test accuracy (Kaggle)

```python
import matplotlib.pyplot as plt

# model names and their corresponding accuracies
model_names = ['VGG16', 'VGG19', 'MobileNetV2', 'ResNet50V2', 'InceptionV3', 'Xception', 'DenseNet201']
accuracies = [73.857, 74.365, 75.888, 76.649, 74.111, 72.842, 72.335] # in percentage

# custom colors for each model
colors = ['#FFB347', '#AEC6CF', '#B19CD9', '#F49AC2', '#C5FB4D', '#F7DC6F', '#85C1E9']

# plot the accuracy values
plt.bar(model_names, accuracies, color=colors)

# add labels to each bar
for i in range(len(model_names)):
    plt.text(i, accuracies[i], str(accuracies[i]), ha='center', va='bottom')

# customize the chart
plt.title('Test Accuracy of Seven Models (Kaggle)')
plt.xlabel('Model Names')
plt.ylabel('Accuracy (%)')
plt.xticks(rotation=45)
plt.ylim([0, 100]) # set the y-axis limits to 100%

# show the plot
plt.show()
```
